# Visualizando ações da Tesla de 2020

```python
# Importando as bibliotecas
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import mplfinance as mpl
```

```python
# Selecionando o período
inicio = dt.datetime(2020,1,1)
fim = dt.datetime(2020,12,31)
```

```python
# Conectando com o Yahoo Finance
df = web.DataReader('TSLA', 'yahoo',inicio, fim)
```

```python
# Visualizando as 10 primeiras linhas
df.head(10)
```

```python
# Salvando o dataset como csv
df.to_csv('TSLA.csv')
```

```python
# Lendo o arquivo csv e visualizando as 10 últimas linhas
acao_tesla = pd.read_csv('TSLA.csv')
acao_tesla.tail(10)
```

```python
# Gráfico de linhas
acao_tesla.plot()
plt.show()
```

```python
# Gráfico Candlestics
mpl.plot(df, type='candle')
plt.show()
```
