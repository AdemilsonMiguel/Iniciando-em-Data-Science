```python
import pandas as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('GIRASSOL.jpg')
```

```python
plt.imshow(img)
plt.xticks([], []) #retirar ticks
plt.yticks([], [])
```

```python
plt.imshow(img[:, :, 2], cmap='gray')
```

```python
img_R = img.copy()
img_R[:, :, [1, 2]] = 0
plt.imshow(img_R)
plt.xticks([], [])
plt.yticks([], [])
```

```python
img_G = img.copy()
img_G[:, :, [0, 2]] = 0
plt.imshow(img_G)
plt.xticks([], [])
plt.yticks([], [])
```

```python
img_B = img.copy()
img_B[:, :, [0, 1]] = 0
plt.imshow(img_B)
plt.xticks([], [])
plt.yticks([], [])
```
