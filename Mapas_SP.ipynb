```python
# Importando as Bibliotecas
import folium
from folium.plugins import HeatMap
import pandas as pd
```

```python
locais = []
```

```python
# Acessando o arquivo excel
ler = pd.read_excel('Mapa_Cidades_SP.xlsx')
```

```python
# É necessário localizar a Latitude e Longitude
mapa = folium.Map(location=[-14.257162, -50.614246],
        zoom_start=5)
```

```python
locais = ler[['LAT', 'LONG']].values.tolist()
```

```python
HeatMap(locais, radius=20).add_to(mapa)
```

```python
# Mapa de calor
mapa
```

```python
# Mapa preto e branco. É possível alterar o zoom com o: ' zoom_start=' podendo modificar este valor para melhorar a precisão do mapa.
mapa2 = folium.Map(location=[-14.257162, -50.614246],
        zoom_start=5, tiles = 'Stamen Toner')
```
