```python
#Importando a biblioteca
import quantstats as qs
```

```python
#Fazendo download dos dados no site Yahoo Finance
petro = qs.utils.download_returns('petr4.sa')
```

```python
#Visualizando os dados dos últimos 21 anos
petro
```

```python
qs.extend_pandas()
```

```python
#Visualizando os retornos mensais
petro.monthly_returns()
```

```python
#Plotando Gráafico 
petro.plot_monthly_heatmap()
```

```python
#Todos os módulos da biblioteca quantstats
dir(qs.stats)
```

```python
#Volatilidade
petro.volatility()
```

```python
#Relatório básico dos retornos
qs.reports.basic(petro, benchmark='^BVSP');
```

```python
#Relatório completo
qs.reports.full(petro, benchmark='^BVSP');
```

```python
#Plotando Gráfico individual
petro.plot_rolling_beta('^BVSP')
```

```python
#Cálculo de retorno anual composto
qs.stats.cagr(petro)
```
